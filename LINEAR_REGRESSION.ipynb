```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn import linear_model

# Load the dataset
df = pd.read_csv("/content/honeyproduction.csv")

# Group data by year and calculate mean total production per year
prod_per_year = df.groupby('year')['totalprod'].mean().reset_index()

# Extracting features (X) and target variable (y)
X = prod_per_year['year'].values.reshape(-1, 1)
y = prod_per_year['totalprod']

# Create subplots
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(12, 5))

# Scatter plot of historical data on the first subplot
axes[0].scatter(X, y)
axes[0].set_xlabel('Year')
axes[0].set_ylabel('Total Production')
axes[0].set_title('Historical Data')

# Linear regression model
regr = linear_model.LinearRegression()
regr.fit(X, y)

# Plotting the regression line on the first subplot
axes[0].plot(X, regr.predict(X), color='red')

# Scatter plot of historical data on the second subplot
axes[1].scatter(X, y)
axes[1].set_xlabel('Year')
axes[1].set_ylabel('Total Production')
axes[1].set_title('Future Prediction')

# Plotting future prediction on the second subplot
X_future = np.array(range(2013, 2050)).reshape(-1, 1)
future_predict = regr.predict(X_future)
axes[1].plot(X_future, future_predict, color='green')

# Show the plots
plt.tight_layout()
plt.show()
```
